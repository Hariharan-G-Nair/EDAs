## Importing the necessary libriaries

```python
df.Dining_Rating=df.Dining_Rating.fillna(df.Dining_Rating.mode())
df.Delivery_Rating=df.Delivery_Rating.fillna(df.Delivery_Rating.mode())
```

## Visualizations

```python
df1=df[(df["Dining_Rating"]>=4.0) & (df["Delivery_Rating"]>=4.0)]
df1.head()
```

```python
df1["Delivery_Votes"].value_counts()	
```

__Top 10 dining restaurant__

```python
df2=df1[(df1["Dining Votes"]>=100)]
dining_restaurant=df2.groupby("Restaurant_Name")["Dining_Rating"].max()
dining_restaurant=pd.DataFrame(dining_restaurant)
best_dining_restaurant=dining_restaurant.sort_values(by=["Dining_Rating"],ascending=False)

plt.figure(figsize=(6,5))
best_dining_restaurant.head(10).plot(kind="bar")
plt.xlabel("Restaurant name")
plt.ylabel("Dining rating")
plt.title("Top 10 Dining Restaurants")
```

__Top 10 delivering restaurant__

```python
df2=df1[(df1["Delivery_Votes"]>=100)]
delivery_restaurant=df2.groupby("Restaurant_Name")["Delivery_Rating"].max()
delivery_restaurant=pd.DataFrame(delivery_restaurant)
best_delivery_restaurant=delivery_restaurant.sort_values(by=["Delivery_Rating"],ascending=False)

plt.figure(figsize=(6,5))
best_delivery_restaurant.head(10).plot(kind="bar")
plt.xlabel("Restaurant name")
plt.ylabel("Delivery rating")
plt.title("Top 10 Delivery Restaurants")


```

__Cities with most popular restaurants__

```python
best_city=df1.groupby("City")["Votes"].sum().nlargest(10)

plt.figure(figsize=(6,5))
best_city.head(10).plot(kind="bar")
plt.xlabel("City")
plt.ylabel("Votes")
plt.title("Top 10 Cities")




```

__Favourite food__

```python
best_items=df1.groupby("Item_Name")["Votes"].sum().nlargest(5)

plt.figure(figsize=(6,5))
plt.pie(best_items,labels=best_items.index)
plt.title("Top 5 favourite food items")
```

__Best sellers__

```python
best_seller=df1.groupby("Best_Seller")["Votes"].sum().nlargest(5)

plt.figure(figsize=(6,5))
plt.pie(best_seller,labels=best_seller.index)
plt.title("Top 5 Best sellers")
```

__Most expensive restaurants__

```python
most_expensive_restaurants=df1.groupby("Restaurant_Name")["Prices"].mean().nlargest(10)

plt.figure(figsize=(6,5))
most_expensive_restaurants.plot(kind="bar")
plt.xlabel("Restaurant name")
plt.ylabel("Average Price")
plt.title("Top 10 most expensive restaurants")

```

__Popular cuisines__

```python
df4=df1[(df1["Dining Votes"]>=100) & (df1["Delivery_Votes"])]
favourite_cuisine=df4.groupby("Cuisine ")[["Dining_Rating","Delivery_Rating"]].sum()	
favourite_cuisine["Total votes"]=favourite_cuisine["Dining_Rating"]+favourite_cuisine["Delivery_Rating"]
most_favourite_cuisine=favourite_cuisine.nlargest(4,"Total votes")

plt.figure(figsize=(6,5))
most_favourite_cuisine.plot(kind="bar")
plt.xlabel("Cuisine")
plt.ylabel("Rating")
plt.title("Most popular Cuisines")
```

__Expensive Cuisines__

```python
expensive_cuisine=df1.groupby("Cuisine ")["Prices"].mean().nlargest(5)

plt.figure(figsize=(6,5))
expensive_cuisine.plot(kind="bar")
plt.xlabel("Cuisine")
plt.ylabel("Average Price")
plt.title("Top 5 most expensive cuisines")


```

__Expensive cities__

```python
expensive_cities=df1.groupby("City")["Prices"].mean().nlargest(5)

plt.figure(figsize=(6,5))
expensive_cities.plot(kind="bar")
plt.xlabel("City")
plt.ylabel("Average Price")
plt.title("Top 5 most Cities")
```
