# Avengers data EDA

## Importing the necessary Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Reading the dataset and data preprocessing

```python
#reading the dataset
df=pd.read_csv(r"C:\Users\HARIHARAN\Downloads\avengers.csv",encoding = "latin-1")
df.head()
```

```python
#removing the column URL from the dataset as it is not necessary for out analysis
df.drop(["URL"],axis=1,inplace=True)
```

```python
df.head()
```

```python
#Enquiring how many rows and columns are there in the dataset
df.shape
```

```python
#Information regarding the dataset
df.info()
```

```python
#All the unique names or aliases used by the avengers
df["Name/Alias"].unique()
```

```python
#Droping any duplicate columns existing in the dataset
df.drop_duplicates()
```

```python
df.info()
```

```python
#Droping the columnn Probationary Introl as it is not necessary for our analysis
df.drop('Probationary Introl',axis=1,inplace=True)
```

```python
#Filling out the null values in all the death and return coulmns as Never Happen as the avengers does'nt die in many or any occasions
df['Death1'].fillna('Never Happen', inplace = True)
df['Death2'].fillna('Never Happen', inplace = True) 
df['Death3'].fillna('Never Happen', inplace = True) 
df['Death4'].fillna('Never Happen', inplace = True) 
df['Death5'].fillna('Never Happen', inplace = True) 
df['Return1'].fillna('Never Happen', inplace = True)
df['Return2'].fillna('Never Happen', inplace = True) 
df['Return3'].fillna('Never Happen', inplace = True) 
df['Return4'].fillna('Never Happen', inplace = True) 
df['Return5'].fillna('Never Happen', inplace = True) 
df["Full/Reserve Avengers Intro"].fillna('Unknown', inplace = True) 
```

```python
#Cleaning the value of the column notes
df.Notes = df.Notes.str.replace("_"," ")
```

```python
df.head()
```

```python
#Viewing the numbers of null values in each column
df.isnull().sum()
```

```python
#Deleting the null values in the colun Name/aliases and reseting index
df = df.dropna(subset=['Name/Alias'])


df.reset_index(drop=True, inplace=True)
```

```python
df.isnull().sum()
```

## Visualizations

```python
#Plotting the distribution of the members of avengers by gender

plt.figure(figsize=(8,5))
sns.countplot(data = df, x="Gender",hue="Honorary",palette='Set1')
plt.xlabel("Gender")
plt.ylabel("Number of members")
plt.xticks(rotation=90)
plt.title("Number of Male and Female Honorary members")
```

##### Accoridng to the graph,number of  of male avengers is far greater than that of female members but the number of avengers from the academy is of similar count from both the genders.

```python
#Polotting a graph to show the numbers of avengers who are currently a part of th eteam or not

plt.figure(figsize=(8,5))
sns.countplot(data = df, x="Current?",hue="Gender",palette='Paired')
plt.xlabel("Current member or not")
plt.ylabel("Number of members")
plt.xticks(rotation=90)
plt.title("Number of Members who are currently serving in the Avengers or not")
```

##### By lookin at the graph we can see that the currently active male members of avengers are a little less that that of the avengers who are not currently a member but the count of female members are almost similar.This means the number of Female members always remained constant throught out the times.

```python
#PLotting a garph to see how many years served by the avengers by gender

plt.figure(figsize=(8,5))
sns.boxplot(y='Years since joining',x='Current?',hue="Gender",data=df)
plt.xlabel("Current member or not")
plt.ylabel("Years since joining")
plt.xticks(rotation=90)
plt.title("Number of years served as an Avenger by Gender"
```

##### From the graph it is clear that the average number of years served by the active female members are higher than that of the male members and not a big differnce is observed on the side of the inactive members.

```python
#Creating two datasets where on one consists of the data about the avnegers who have died and other one about the avenders that never died

still_alive = df[df['Notes'].isnull()]
death = df[df['Notes'].notnull()]
```

```python
still_alive.head()
```

```python
death.head()
```

```python
##Plotting the death toll of avengers by gender

plt.figure(figsize=(8,5))
sns.countplot(data = death, x="Gender",palette='coolwarm')
plt.xlabel("Gender")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.title("count of Avengers who died")
```

##### The death toll of male members are shown to be drastically higher that that of the female members.

```python
##PLotting the number of avnegers who never dies with respect to gender

plt.figure(figsize=(8,5))
sns.countplot(data = still_alive, x="Gender",palette='coolwarm')
plt.xlabel("Gender")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.title("count of Avengers who are not dead")
```

##### The survival rate of male members are also shown to be higher than that of the female members.
