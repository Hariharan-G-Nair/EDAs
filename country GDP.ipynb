```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\HARIHARAN\Downloads\API_NY.GDP.MKTP.CD_DS2_en_csv_v2_26\API_NY.GDP.MKTP.CD_DS2_en_csv_v2_26.csv")
df.head()
```

```python
df = df.dropna(subset=['Country Name'])
df = df.reset_index(drop=True)
```

```python
df.drop(["Indicator Code","Indicator Name","Country Code","2023"],axis=1,inplace=True)
```

```python
df.head()
```

```python
df["Country Name"].unique()
```

```python
df.info()
```

```python
df1=df.bfill(axis="columns")
df1
```

```python
df2=df1.ffill(axis="columns")
df2.head()
```

```python
df2.info()
```

```python
for col in df2.columns[1:]:
    df2[col] = pd.to_numeric(df2[col], errors='coerce')
```

```python
df2.info()
```

```python
df2["GDP of last 10 years"]=df2[["2013","2014","2015","2016","2017","2018","2019","2020","2021","2022"]].sum(axis=1)
df2.head()
```

```python
df3=df2.sort_values(by=["GDP of last 10 years"],ascending=False)
df3.head()
```

```python
plt.figure(figsize=(8, 6))
plt.bar(df3["Country Name"].head(),df3["GDP of last 10 years"].head())
plt.xlabel("Country name")
plt.ylabel("GDP in the last 10 years in trillions")
plt.xticks(rotation=90)
plt.title("Top 5 countries with the most GDP in last 10 years")
```

```python
plt.figure(figsize=(8, 6))
plt.bar(df3["Country Name"].tail(),df3["GDP of last 10 years"].tail())
plt.xlabel("Country name")
plt.ylabel("GDP in the last 10 years in trillions")
plt.xticks(rotation=90)
plt.title("top 5 countries with the least GDP in last 10 years")
```

```python
df3.tail()
```

```python
plt.figure(figsize=(12, 6))
for country in df3['Country Name']:
    plt.plot(df3.columns[1:-1], df3[df3['Country Name'] == country].values[0][1:-1], label=country)
plt.title('GDP Trends Over Time')
plt.xlabel('Year')
plt.ylabel('GDP (in Trillions)')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
def visualize_gdp_trend(df, countries):
    
    plt.figure(figsize=(12, 6))
    for country in countries:
        gdp_values = df[df['Country Name'] == country].values[0][1:-1]  
        plt.plot(df.columns[1:-1], gdp_values, label=country)
    plt.title('GDP Trends Across Years')
    plt.xlabel('Year')
    plt.ylabel('GDP (in Trillions)')
    plt.xticks(rotation=90)
    plt.legend()
    plt.show()
```

```python
visualize_gdp_trend(df,["United States"])
```

```python
visualize_gdp_trend(df,["India"])
```

```python
visualize_gdp_trend(df,["China"])
```
